```python
nuisance_params_14=['trans_x', 'trans_x_derivative1', 'trans_y', 'trans_y_derivative1', 'trans_z', 'trans_z_derivative1', 
                    'rot_x', 'rot_x_derivative1', 'rot_y', 'rot_y_derivative1', 'rot_z', 'rot_z_derivative1', 'csf', 'white_matter']
motion_outliers=['motion_outlier_*']


with open('nuisance_params_14.txt', 'w') as f: 
    f.write(str(nuisance_params_14))
```

```python
df = pd.read_csv('/Users/adamkrause/TIRED/derivatives/fmriprep-22.1.1/sub-TIRED031/ses-BL/func/sub-TIRED031_ses-BL_task-con2_desc-confounds_timeseries.tsv', sep='\t')
df.head()
```

```python
## load in the nuisance_params list to filter the confounds.tsv file
with open('nuisance_params_14.txt', 'r') as f: 
    list_str = f.read()

nuisance_params_14 = eval(list_str)
```

```python
nuisance_cols = df.loc[:, nuisance_params_14].copy()
motion_cols = df.filter(regex='^motion_outlier_*').copy()
motion_cols.shape[0]
```

```python
for col in motion_cols.columns:
    # find the index of the first 1 in the column
    index = motion_cols[col].idxmax()
    # set the previous and next two values to 1, if they exist
    if index > 0:
        motion_cols[col].iloc[index-1:index+3] = 1
    elif index == 0:
        motion_cols[col].iloc[:3] = 1
    elif index == len(df)-1:
        motion_cols[col].iloc[-3:] = 1

motion_cols.head()
    
    
    
    

            
```

```python
## count the number of volumes marked as spikes after -1/2+ adjustment
tempdata = {'Column 1': [1, 1, 1, 0, 0], 
        'Column 2': [0, 1, 1, 1, 0], 
        'Column 3': [0, 0, 1, 1, 1]}
tempdf = pd.DataFrame(tempdata)

tempdf
```

```python
badvols = set()
for row in tempdf.itertuples(index=False):
    if 1 in row:
        badvols.add(row)
percent = round((len(badvols) / tempdf.shape[0]) * 100 ,1)
filename = f"OutlierVols_total-{len(badvols)}_perc-{percent}.txt"
with open(filename, "w") as f:
    f.write(("Total outlier volumes: " + str(len(badvols)) + "\nPercent outlier volumes: " + str(percent)))
print(f"Outlier Volume Info saved to {filename}")
```

```python
denoise_regressors_df = pd.concat([nuisance_cols, motion_cols], axis=1)
denoise_regressors_df = denoise_regressors_df[3:]

denoise_regressors_df.shape
```

```python
data=denoise_regressors_df.values
savemat('data.mat', {'data': data})
```
